```python
import cv2
import matplotlib.pyplot as plt

# Change this path to your image location in Drive
image_path = "/content/drive/MyDrive/image.jpg"

# Read image
img = cv2.imread(image_path)

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.title("Grayscale Image")
plt.show()
```

```python
# Sobel horizontal edges
sobel_h = cv2.Sobel(gray, cv2.CV_64F, dx=0, dy=1, ksize=3)

# Sobel vertical edges
sobel_v = cv2.Sobel(gray, cv2.CV_64F, dx=1, dy=0, ksize=3)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(sobel_h, cmap='gray')
plt.title("Horizontal Edges")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sobel_v, cmap='gray')
plt.title("Vertical Edges")
plt.axis("off")

plt.show()
```

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.title("Canny Edges")
plt.show()
```

```python
import numpy as np

# Input matrix
A = np.array([
    [0,0,1,1,0,0,0,1,1],
]*12)  # 12 rows

# Edge detection: column differences
edge_binary = np.zeros((A.shape[0], A.shape[1]-1))

for j in range(A.shape[1]-1):
    edge_binary[:, j] = (A[:, j] != A[:, j+1]).astype(int)

print("Binary Edge Map (distinct edges):")
print(edge_binary)
```
